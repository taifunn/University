0. Znajdź i pobierz zbiór danych CAR EVALUATION z repozytorium danych UCI. Dowiedz się
co opisują pobrane dane.

```python
import pandas as pd
import matplotlib.pyplot as plt

cols = ['buying', 'maint','doors','persons','lug_boot','safety', 'class']
path = "car.data"
df = pd.read_csv(path, header=None, names=cols)
```

1. Na wykresie tortowym pokaż ile spośród wszystkich samochodów to samochody nieakceptowalne, akceptowalne, dobre i bardzo dobre.

```python
vals = df['class'].value_counts().reindex(['unacc','acc','good','vgood'], fill_value=0)

plt.figure()
plt.pie(
    vals.values,          
    labels=vals.index, 
    autopct='%.1f%%'
)
plt.axis('equal') 
plt.show()
```

2. Na wykresie słupkowym pokaż ile jest samochodów 2, 3 i 4 drzwiowych w różnych klasach
bezpieczeństwa.

```python
import pandas as pd
import matplotlib.pyplot as plt

sub = df[df['doors'].isin(['2','3','4'])]

counts = pd.crosstab(sub['safety'], sub['doors'])

counts[['2','3','4']].loc[['low','med','high']].plot.bar()

plt.xlabel('bezpieczenstwo')
plt.ylabel('liczba aut')
plt.title('Auta 2/3/4-drzwiowe w klasach bezpieczeństwa')
plt.xticks(rotation=0)
plt.legend(title='drzwi')
plt.show()
```
