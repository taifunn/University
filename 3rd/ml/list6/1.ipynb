```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

X = np.array([
    [0, 0],
    [0, 1000],
    [0, 2000],
    [10, 0],
    [10, 1000],
    [10, 2000],
])

kmeans_raw = KMeans(n_clusters=2, random_state=0, n_init=10)
labels_raw = kmeans_raw.fit_predict(X)

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

print("Dane przed standaryzacjÄ…:")
print(X)
print("\nDane po standaryzacji:")
print(X_std)

kmeans_std = KMeans(n_clusters=2, random_state=0, n_init=10)
labels_std = kmeans_std.fit_predict(X_std)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(X[:, 0], X[:, 1], c=labels_raw)
axes[0].set_title("KMeans bez standaryzacji")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].grid(True)

axes[1].scatter(X[:, 0], X[:, 1], c=labels_std)
axes[1].set_title("KMeans po standaryzacji")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
