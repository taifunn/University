```python
import pandas as pd 
import matplotlib.pyplot as plt

path = "data4.csv"
df = pd.read_csv(path, sep=',')

df.head(10)
```

```python
df.describe()
```

```python
df.hist(bins=30, figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
def correlation(input, output, input_name, output_name, title):
    plt.figure(figsize=(6,4))
    plt.scatter(input, output)
    plt.xlabel(input_name)
    plt.ylabel(output_name)
    plt.title(title)

features = [f"x{i}" for i in range(1, 20)]
for f in features:
    correlation(df[f], df["y"], f, "y", f"zależność {f} i y")
```

- x8 i x9 maja najmocniejsza zaleznosc liniowa z y
- nie sa wspoliniowe
- wartosci x sa dosc podobne do siebie

```python
import numpy as np
from sklearn.model_selection import train_test_split

xsy = [f"x{i}" for i in range(1, 20)] 
X = df[xsy]
Y = df['y']

x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, shuffle=True, random_state=0
)
```

Model baselinowy:

```python
from sklearn.dummy import DummyRegressor

dum = DummyRegressor(strategy="mean").fit(x_train, y_train)
y_pred = dum.predict(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    mean_squared_error, r2_score
)

def preprocess_data(x_train, y_train):
    scal = StandardScaler()
    x_proc = scal.fit_transform(x_train)

    return x_proc, y_train, scal 
```

Regresja:

```python
x_train, y_train, pre = preprocess_data(x_train, y_train) 
lin = LinearRegression().fit(x_train, y_train)
x_test = pre.transform(x_test)
y_pred = lin.predict(x_test)

mse_lin = mean_squared_error(y_test, y_pred)
r2_lin = r2_score(y_test, y_pred)
```

Regularyzacja:

```python
from sklearn.linear_model import RidgeCV

x_train, y_train, pre = preprocess_data(x_train, y_train) 
reg = RidgeCV(alphas=np.logspace(-8, 3, 60), cv=5)
reg.fit(x_train, y_train)
x_test = pre.transform(x_test)
y_pred = reg.predict(x_test)

mse_reg = mean_squared_error(y_test, y_pred)
r2_reg = r2_score(y_test, y_pred)
```

```python
results = pd.DataFrame([
    {"Model": "Regresja", "MSE": mse_lin, "R2": r2_lin},
    {"Model": "Regularyzacja", "MSE": mse_reg, "R2": r2_lin},
])

print(results)
```

Wnioski:
- Oba działają podobnie, bo x nie są współliniowe (gdyby były drugi byłby znacznie lepszy od pierwszego),
poniewaz RidgeCV przy dobrych warunkach danych zachowuje sie jak zwykla regresja.
